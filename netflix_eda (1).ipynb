This project explores trends and patterns in Netflix movies and TV shows using Python-based data analysis and visualization.


## Problem Statement

Streaming platforms generate large volumes of content data.  
This project aims to analyze Netflix content to understand:
- The distribution of Movies vs TV Shows
- Trends in content release over time
- Patterns that indicate growth in content production


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df['type'].value_counts()
```

```python
df['release_year'].value_counts().head(10)
```

```python
df['type'].value_counts().plot(
    kind='bar',
    figsize=(6,4)
)
plt.title("Distribution of Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()



```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Released Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
top_countries = df['production_countries'].value_counts().head(10)
top_countries
```

```python
top_countries.plot(kind='bar', figsize=(8,4))
plt.title("Top 10 Content Producing Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()

```

## Insights
- Movies dominate Netflixâ€™s content library compared to TV Shows.
- Content production increased sharply after 2015, indicating platform expansion.
